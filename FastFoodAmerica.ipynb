```python
import pandas as pd
import sweetviz as sv
```

```python
import pandas as pd

# Cargar los DataFrames
df1 = pd.read_csv("Datafiniti_Fast_Food_Restaurants.csv")
df2 = pd.read_csv("Datafiniti_Fast_Food_Restaurants_May19.csv")

# Obtener listas de columnas
columns_df1 = df1.columns.tolist()
columns_df2 = df2.columns.tolist()

# Imprimir información sobre los DataFrames
df1.info()
df2.info()
```

```python
# Comparar las listas de columnas
compare_and_print = lambda columns_df1, columns_df2: print("Se pueden concatenar los DataFrames, ya que comparten las mismas columnas.") if columns_df1 == columns_df2 else print("Es necesario aplicar merge, ya que los DataFrames tienen columnas diferentes.")

# Luego puedes llamar a la función pasando las listas de columnas como argumentos
compare_and_print(columns_df1, columns_df2)

# Definir la función para identificar columnas diferentes
identify_and_print_diff_columns = lambda columns_df1, columns_df2: print(f"Las columnas diferentes son: {set(columns_df1) ^ set(columns_df2)}") if columns_df1 != columns_df2 else print("Se pueden concatenar los DataFrames, ya que comparten las mismas columnas.")

# Luego puedes llamar a la función pasando las listas de columnas como argumentos
identify_and_print_diff_columns(columns_df1, columns_df2)
```

```python

# Concatenar los DataFrames verticalmente (uno encima del otro)
concatenated_df = pd.concat([df1, df2], ignore_index=True)
```

```python
concatenated_df.info()
```

```python
# Crear informe comparativo
report = sv.compare([df1, "DF1"], [df2, "DF2"])

# Guardar el informe en un archivo HTML
report.show_html("sweetviz_report.html")
```

```python
report.show_notebook()
```

```python
# Crear un informe para un solo DataFrame
report = sv.analyze([df1, 'DF1'])

# Guardar el informe en un archivo HTML
report.show_html('single_dataframe_report.html')
```
